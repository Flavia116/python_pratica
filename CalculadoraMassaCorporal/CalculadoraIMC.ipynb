```python
print("\nCalculadora  índice de massa corporal (IMC) >>> \n")

#entrada
nome =   input("\nQUAL SEU NOME: ").upper()
peso =   float(input("\nQUAL SEU PESO: "))
altura = float(input("\nQUAL SUA ALTURA: "))

print("\nClassificação Grau de Obesidade >>>> \n")

Magreza = 18.5
Normal = 18.5 and 24.9	
SobrepesoI =  25.0 and 29.9	
ObesidadeIII = 30.0 and 34.9	
ObesidadeIIII = 35.0 and 39.9	
ObesidadeGraveIII =  40.0	

#processamento
imc = peso/(altura * altura)

if imc < 18.5:
        resultado = "Magreza 😔"
    
elif imc < 25:
        resultado = "Normal 🥳 "

elif imc < 30 :
        resultado = "Sobrepeso I 🥺 "
    
elif imc < 35 :
        resultado = "Obesidade III 😢"
    
elif imc < 40 :
        resultado = "Obesidade IIII 😭"

else:
        resultado = "ObesidadeGraveIII 😭😭"

#saída
print(f"Olá, {nome}! Seja bem-vinda!  Seu índice de massa corporal é: {imc:0.2f} == {resultado} ")
```
