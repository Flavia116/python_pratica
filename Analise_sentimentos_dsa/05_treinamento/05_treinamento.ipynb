```python
criterion = nn.CrossEntropyLoss()

optimizer = optim.SGD(
    modelo_dsa.parameters(),
    lr=learning_rate,
    momentum=0.9
)

print("\nIniciando treinamento...\n")

n_total_steps = len(dsa_loarders_treino)

for epoch in range(num_epochs):

    modelo_dsa.train()
    running_loss = 0.0

    for imagens, labels in dsa_loarders_treino:

        imagens = imagens.to(device)
        labels = labels.to(device)

        outputs = modelo_dsa(imagens)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Avaliação
    modelo_dsa.eval()
    n_correct = 0
    n_samples = 0

    with torch.no_grad():
        for val_images, val_labels in dsa_loader_teste:

            val_images = val_images.to(device)
            val_labels = val_labels.to(device)

            val_outputs = modelo_dsa(val_images)
            _, predicted = torch.max(val_outputs.data, 1)

            n_samples += val_labels.size(0)
            n_correct += (predicted == val_labels).sum().item()

    acc = 100.0 * n_correct / n_samples
    avg_loss = running_loss / n_total_steps

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}, Acc: {acc:.2f}%")
```
