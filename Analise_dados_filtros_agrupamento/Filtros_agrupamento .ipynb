```python
import numpy as np
import pandas as pd
```

```python
#01 A parti do DataFrame de funcionário abaixo, selecione  e exiba os funcionários do departamento de "vendas" que tem 
# um salário superior R$ 4500

dados = {"nome": ["Ana", "Bruno", "Carla", "Matias", "Eliana", "Fabiano"],
        "Departamento": ["RH", "Vendas", "TI", "Vendas", "RH", "Vendas"],
        "Salário": [4000, 5000, 6200, 4400, 4300,5500]}

df_funcionários = pd.DataFrame(dados)

df_funcionários

condicao_depto = df_funcionários["Departamento"] == "Vendas"

condicao_salairo = df_funcionários["Salário"] > 4500

funcionarios_filtrados = df_funcionários[condicao_depto & condicao_salairo]

funcionarios_filtrados
```

```python
#02 Com o DataFrame de vendas abaixo, calcule o total de vendas(Valor) para cada categoria de produto.

data_vendas = { "Categoria": ["Eletronicos",  "Vestuário", "Eletronicos", "Casa", "Vestuário"],
                "Produto": ["TV", "Camiseta", "Notebook", "Sofá", "Calça",],
                "Valor": [2500,80, 4500, 3000, 1000]}

data_vendas1 = pd.DataFrame(data_vendas)

data_vendas1

vendas_por_categoria = data_vendas1.groupby("Categoria")["Valor"].sum()

print(vendas_por_categoria)
```

```python
#03 A parti do DataFrame de produtos abaixo, crie uma nova coluna chamada preco_com_Desconto que contenha o 
#valor da coluna Preco com 10% desconto

dadosProdutos = {"Produto": ["Monitor", "Teclado", "Mouse", "Webcam"],
                "Preco": [800, 120,70,250]}

df_produtos = pd.DataFrame(dadosProdutos)

df_produtos

df_produtos["Preco_com_Desconto"] = df_produtos["Preco"]* 0.90

df_produtos
```

```python
#04 No DataFrame de alunos abaixo, subustitua as notas ausentes(NAN) pela média da turma (média da coluna "Nota")

dados_alunos = {"Aluno" : ["Aline", "Bernardo", "Clara", "Marcelo"],
                "Nota" : [8.5, 7.0, np.nan, 9.0]}

df_alunos = pd.DataFrame(dados_alunos)

df_alunos

media_notas = df_alunos["Nota"].mean()

df_alunos[ "Nota"] = df_alunos["Nota"].fillna(media_notas)

df_alunos
```

```python
#05 Ordene o DataFrame de pontuações abaixo, do maior para o menor com base na coluna "pontos e exiba"
# o resultado. 

dados_pontuacao = {"Dados": ["J1", "J2", "J3", "J4", "J5"],
                    "Pontos": [88,95,74,102,95]}

dados_pontuacao = pd.DataFrame(dados_pontuacao)

dados_pontuacao 

df_ordenado = dados_pontuacao.sort_values(by = "Pontos", ascending = False)

df_ordenado
```

```python
#06 Combine os DataFrame df_pedidos com base na coluna em comum ID_Cliente

df_cliente = pd.DataFrame({"ID_Cliente": [1,2,3],
                            "Nome": ["Carlos", "Mariana", "Lucas"]})

df_pedidos = pd.DataFrame({"ID_Pedido": [101,102,103],
                           "ID_Cliente": [2,1,2],
                            "Produto": ["Livro","Caneta", "Caderno"]})

df_combinado = pd.merge(df_cliente,df_pedidos, on = "ID_Cliente")

df_combinado
```

```python
#07 No DataFrame abaixo, extraia o ano da coluna Data e Crie uma nova coluna chamada Ano. 

dados_eventos = {"Evento": ["Conferência A", "Workshop B", "Feira C"],
                "Data": ["2025-10-25", "2026-03-12", "2026-09-01"]}

df_eventos = pd.DataFrame(dados_eventos)

df_eventos

df_eventos.info()

df_eventos["Ano"] = df_eventos["Data"].str.slice(0,4)

df_eventos

df_eventos["Ano"] = pd.to_datetime(df_eventos["Data"]).dt.year

df_eventos

df_eventos["Ano"] = pd.to_datetime(df_eventos["Data"]).dt.year

df_eventos
```

```python
#08 Crie uma nova coluna Status no DataFrame de notas.  Se a nota for maior ou igual 7, O status
#deve ser "APROVADO", caso contrário, "REPROVADO" e use o metodo apply

dados_notas = {"Aluno": ["Maria", "Jeremias", "Paulo", "Roberto"],
              "Nota": [9.5,6.0,5.5,8.0]}

df_notas = pd.DataFrame(dados_notas)

df_notas

df_notas["Status"] = df_notas["Nota"].apply(lambda nota: "Aprovado" if nota >= 7 else "Reprovado")

df_notas
```

```python
#09 Crie uma tabela dinâmica que mostre a soma das vendas por Religião e Vendedor. 

dados_regional = {"Regiao": ["Norte", "Sul", "Norte", "Sul", "Norte", "Sul"],
                 "Vendedor":["Ana", "Bruno", "Gabriela", "Carlos", "Bruno", "Paulo"],
                 "Vendas": [1000, 1500, 1200, 1800, 800, 1300]}


df_regional = pd.DataFrame(dados_regional)

df_regional

tabela_dinamica = df_regional.pivot_table(index = "Regiao",
                                          columns = "Vendedor", 
                                          values = "Vendas",
                                          aggfunc = "sum",
                                          fill_value = 0)

tabela_dinamica
```
