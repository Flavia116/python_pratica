```python
#Verificador de Número Par ou Ímpar
Este projeto em Python solicita o nome do usuário e um número inteiro, informando se o número digitado é par ou ímpar.
```

```python
#Entrada 
try:
    nome = input("Informe seu nome: \n")
    numero2 = int(input("Digite um número: \n"))

#Processamento
    if numero2 % 2 == 0:
        resultado= "par"

    else:
        resultado= "ímpar"

#Saída
    print(f" Olá, {nome}. Esperamos que esteja bem! O número digitado foi: {numero2} é {resultado}." )

except ValueError:
    print("Digite um número válido")       
```
