# Este projeto em Python calcula a média de um aluno com base em 4 provas (P1, P2, P3 e P4) 
# e informa se ele foi aprovado ou reprovado.

```python
#entrada
nome =  input(" Insira seu nome: ").upper()

Nota1 = float((input("informe a nota 1: ")))
Nota2 = float((input("informe a nota 2: ")))
Nota3 = float((input("informe a nota 3: ")))
Nota4 = float((input("informe a nota 4: ")))

#processamento 
somaNotas = (Nota1 + Nota2 + Nota3 + Nota4)/4

#med_notas = somaNotas / 4

if somaNotas >= 7:
    resp = "APROVADO! Parabéns 🎉"

else:
    resp = "REPROVADO 😔"

#saída

print(f"\n Olá estudante, {nome}! Média {somaNotas:0.0f}. Aluno {resp} " )

#fim algoritmo 
```
