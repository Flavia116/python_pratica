```python
#purê de batatas

nome = input(" insira seu nome: ").upper()

geladeira = 5
receita_pedi = 10

Z = receita_pedi - geladeira

if Z > 0:
    resp = "ir ao mercado comprar"

else:
    resp = "não precisa ir comprar"

print(f" Olá, {nome}. Você atualmente tem {geladeira} batatas na geladeira, então {resp} o restante: {Z:0.0f}.")
```
